```python
import opendatasets as od
import pandas as pd
import matplotlib.pyplot as plt
```

# Preparing and Cleaning Data

## Importing datasets from Kaggle

```python
od.download("https://www.kaggle.com/datasets/ilyaskaddouri/electric-cars-vs-motor-cars")
od.download("https://www.kaggle.com/datasets/iannjuguna/electric-car-prices")
```

```python
data1 = pd.read_csv("electric-car-prices/Electric cars.csv")
data1
```

```python
data2 = pd.read_csv("electric-cars-vs-motor-cars/Electric Cars VS Motor Cars.csv", sep = ';')
data2
```

## Cleaning data

From the previous output, we can see that the first data is messy with several unused columns and rows. The data will be cleaned from those problems.

```python
# Cleaning the first data

# Taking the first 13 rows
data1 = data1.iloc[:13]
# Dropping unused columns
data1 = data1.drop(columns=["Unnamed: 5", "Unnamed: 6"])

data1
```

# Data Visualization
* (*) Visualisasi tambahan sebagai bahan menjawab problem yang diberikan
* (**) Visualisasi tambahan sebagai bahan eksplorasi pemahaman data, namun tidak dipakai secara eksplisit untuk menjawab problem yang ada

## Price comparison of motor and EV cars
(Perbandingan biaya mobil BBM dan mobil listrik)


```python
# mengambil kolom data yang dibutuhkan
BEV = data1['BEV average price (USD)']
Motor = data1['Average price of new car']

# mendesain grafik
plt.figure(figsize=(20,8), dpi=200).set_facecolor("azure")
plt.title("Average Price")

# menggambar grafik
plt.plot(data1.Year, BEV, ".-", label="BEV")
plt.plot(data1.Year, Motor, ".-", label="Motor Car")

# label
plt.ylabel("USD", fontsize="x-large")
plt.xlabel("Year", fontsize="x-large")
plt.legend(loc="upper left", fontsize="x-large")

plt.yticks([35000, 40000, 45000, 50000, 55000, 60000, 65000])
plt.grid(True, linestyle="--", color="gainsboro")

plt.show()
```

## Average CO2e emission comparison of motor cars usage and EV cars production
(Perbandingan emisi gas rumah kaca dari pemakaian mobil BBM dan produksi mobil listrik)

```python
# mengambil kolom data yang diperlukan
BEV = data2['Average CO2 emission per BEV produced (Kg CO2e)']
PHEV = data2['Average CO2 emission per PHEV produced (Kg CO2e)']
Motor = data2['Average CO2 emission per Moto car']

# mendesain grafik
plt.figure(figsize=(20,8), dpi=200).set_facecolor("azure")
plt.title("Average CO2e Emission")


# menggambar grafik
plt.plot(data2.Year, BEV, ".-", label="Per Production of BEV")
plt.plot(data2.Year, PHEV, ".-", label="Per Production of PHEV")
plt.plot(data2.Year, Motor, ".-", label="Motor Car")

# label
plt.ylabel("Kg CO2e", fontsize="x-large")
plt.xlabel("Year", fontsize="x-large")
plt.legend(loc="upper left", fontsize="x-large")

plt.grid(True, linestyle="--", color="gainsboro")

plt.show()
```

## Oil displacement of EV cars*
(Substitusi minyak bumi oleh kendaraan listrik*)

```python
# mendesain grafik
plt.figure(figsize=(20,8), dpi=200).set_facecolor("azure")
plt.title("Oil Displacement by Electric Cars per Day")

# menggambar grafik
plt.plot(data2.Year, data2['Oil displacement (Milion litres gasoline equivalent)'], ".-")

# label
plt.ylabel("Million litres", fontsize="x-large")
plt.xlabel("Year", fontsize="x-large")

plt.grid(True, linestyle="--", color="gainsboro")

plt.show()
```

## Sales of EV cars (BEV and PHEV)*
(Angka penjualan mobil listrik berjenis BEV dan PHEV*)


```python
# mengambil kolom data yang diperlukan
BEV = data2['EV sales BEV']
PHEV = data2['EV sales PHEV']

# mendesain grafik
plt.figure(figsize=(20,8), dpi=200).set_facecolor("azure")
plt.title("Sales of Electric Cars")


# menggambar grafik
plt.plot(data2.Year, BEV, ".-", label="BEV")
plt.plot(data2.Year, PHEV, ".-", label="PHEV")

# label
plt.ylabel("Unit Car", fontsize="x-large")
plt.xlabel("Year", fontsize="x-large")
plt.legend(loc="upper left", fontsize="x-large")

plt.grid(True, linestyle="--", color="gainsboro")

plt.show()
```
